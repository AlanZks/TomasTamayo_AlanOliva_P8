## Introducción

En este informe se desarrolla y compara dos enfoques algorítmicos para la detección de patrones en archivos de texto plano: una estrategia de fuerza bruta y otra basada en la construcción de un índice mediante una tabla hash propia. Ambas técnicas se implementan con el objetivo de localizar palabras o frases específicas de manera eficiente en distintos escenarios de búsqueda.

Los métodos son evaluados experimentalmente para analizar su rendimiento, especialmente en términos de tiempo de ejecución. Los resultados obtenidos buscan evidenciar las ventajas de la búsqueda indexada frente a la fuerza bruta, aportando criterios para seleccionar el algoritmo más adecuado según la naturaleza y el volumen de los datos. Esta comparación es especialmente relevante en aplicaciones que procesan grandes volúmenes de información textual, donde la eficiencia puede impactar directamente en la usabilidad y escalabilidad del sistema.

## 1. Análisis del problema

El problema se centra en la necesidad de localizar eficientemente palabras o frases específicas dentro de archivos de texto plano, una tarea común en aplicaciones como análisis de datos,  recuperación de información e indexación.

Se consideran dos enfoques principales:

- **Búsqueda por fuerza bruta**: Consiste en recorrer secuencialmente cada línea y comparar cada posición con el patrón buscado. Si bien garantiza encontrar todas las ocurrencias, su eficiencia disminuye drásticamente en archivos grandes o con múltiples consultas.

- **Búsqueda indexada con tabla hash**: Se construye un índice que asocia palabras a sus ubicaciones, permitiendo acceder rápidamente a los posibles lugares del patrón. Esto reduce el tiempo de búsqueda, pero requiere procesamiento y memoria adicionales para construir y mantener la tabla hash, especialmente en textos con vocabulario extenso.

La elección del método óptimo depende de factores como el tamaño y la complejidad del archivo, la frecuencia de consultas y los recursos computacionales disponibles. Por ejemplo, en archivos pequeños o con consultas esporádicas, la fuerza bruta puede ser suficiente, mientras que en grandes volúmenes de datos o consultas frecuentes, la búsqueda indexada se vuelve más eficiente.

En este análisis, se parte del supuesto de que los archivos son de texto plano, con palabras separadas por espacios, y se evalúan ambos métodos considerando escenarios extremos y limitaciones de recursos.

## 2. Solución del problema

### 2.1. Algoritmo de Solución

Para resolver la problemática de búsqueda eficiente de patrones en un texto, se diseñaron e implementaron dos enfoques complementarios: la búsqueda por fuerza bruta y la búsqueda indexada. Cada uno de estos métodos ofrece ventajas particulares en cuanto a simplicidad, velocidad y escalabilidad, y su integración permite comparar y seleccionar la mejor estrategia según las necesidades del usuario y el tamaño del texto.

En el caso de la **búsqueda por fuerza bruta**, se emplea una estrategia clásica donde el patrón ingresado es buscado exhaustivamente a lo largo de todo el texto. Este procedimiento se realiza línea por línea, comparando cada posible subcadena con el patrón, carácter por carácter. Además, se verifica que las coincidencias identificadas correspondan a palabras exactas, utilizando una función auxiliar que distingue entre caracteres válidos y separadores. Aunque este método es directo y sencillo de implementar, su eficiencia disminuye notablemente en textos de gran tamaño, ya que requiere recorrer todo el contenido sin optimización previa.

Por otro lado, la **búsqueda indexada** se apoya en la construcción previa de un índice de palabras, desarrollado a través de un módulo denominado "Indexador". Este índice se implementa mediante una tabla hash, donde cada palabra del texto es asociada a las líneas en las que aparece. Durante el proceso de indexación, el texto es analizado línea por línea, separando y limpiando las palabras para evitar distorsiones por signos de puntuación o variantes ortográficas. Gracias a este índice, la búsqueda de patrones se agiliza considerablemente: en vez de revisar todo el texto, se consultan únicamente las líneas relevantes, y si el patrón consta de varias palabras, se intersectan los resultados para encontrar coincidencias exactas. Este enfoque es especialmente eficiente cuando se realizan múltiples búsquedas sobre textos extensos.

En resumen, mediante la integración de ambos métodos, se proporciona una solución flexible y eficiente para la búsqueda de patrones en texto, permitiendo alternar entre simplicidad y rapidez según el caso de uso. La arquitectura desarrollada facilita la comparación de resultados y el análisis de rendimiento de cada estrategia.

### 2.3. Implementación

<h3 style="text-align: center;">Buscador Fuerza Bruta</h3>


El pseudocódigo presentado describe el funcionamiento de un buscador de fuerza bruta diseñado para localizar ocurrencias exactas de una palabra o patrón dentro de un texto.

Primero, se define una función encargada de determinar si un carácter es considerado un separador, es decir, si no corresponde a una letra o a un dígito numérico. Esta función resulta fundamental para asegurar que las coincidencias identificadas correspondan a palabras completas y no a fragmentos dentro de otras palabras.

La función principal de búsqueda recorre el texto línea por línea. Para cada línea, se examinan todas las posiciones posibles donde el patrón podría coincidir. En cada posición, se compara carácter por carácter con el patrón buscado. Si se encuentra una coincidencia exacta, se verifica que la coincidencia esté delimitada por separadores, tanto antes como después, para garantizar que se trata de una palabra aislada. Si se cumplen estas condiciones, la función almacena la línea y la posición de la coincidencia.

Finalmente, se retorna una lista con todas las posiciones en las que se identificó el patrón como una palabra exacta dentro del texto.

<pre>
Función es_separador(caracter)
    Si caracter NO es letra, número, ni vocal acentuada
        Retornar Verdadero
    Sino
        Retornar Falso
FinFunción

Función buscar(texto, patrón)
    resultado ← lista vacía
    largo_patron ← longitud(patrón)

    Para cada línea (índice i) en texto
        Para cada posición pos desde 0 hasta (longitud(línea) − largo_patron)
            coincide ← Verdadero

            Para k desde 0 hasta largo_patron − 1
                Si texto[i][pos + k] ≠ patrón[k]
                    coincide ← Falso
                    Salir del ciclo interno
                FinSi
            FinPara

            Si coincide
                antes ← (pos == 0) O es_separador(texto[i][pos − 1])
                después ← (pos + largo_patron == longitud(línea)) O es_separador(texto[i][pos + largo_patron])
                Si antes Y después
                    Agregar (i + 1, pos) a resultado
                FinSi
            FinSi
        FinPara
    FinPara

    Retornar resultado
FinFunción
</pre>






<h3 style="text-align: center;">Buscador Indexado</h3>

El pseudocódigo presentado describe un algoritmo de búsqueda indexada que, a partir de un índice previamente construido, identifica eficientemente las líneas candidatas donde podría encontrarse el patrón.

El patrón es dividido en palabras, y para cada una se obtienen las líneas en las que aparece. Luego se calcula la intersección de estas listas, determinando así las líneas que contienen todas las palabras del patrón.

A continuación, solo en estas líneas candidatas, se verifica carácter por carácter la presencia exacta del patrón completo, asegurando además que esté delimitado por separadores para confirmar coincidencias de palabras exactas.

El algoritmo retorna una lista ordenada con las posiciones de todas las apariciones válidas, permitiendo una búsqueda precisa y altamente eficiente incluso sobre textos extensos.
<pre>
Función es_separador(caracter)
    Retornar Verdadero si el caracter NO es:
        - una letra (mayúscula/minúscula, incluyendo acentos y ñ)
        - un número
    (Es decir, retorna Verdadero si es un separador o símbolo)
FinFunción

Función buscar(indexador, texto, patrón)
    palabras ← lista vacía
    palabra ← cadena vacía

    // Separar el patrón en palabras usando espacios
    Para cada carácter en (patrón + carácter final especial)
        Si el carácter es espacio O es el carácter final
            Si palabra no está vacía
                Agregar palabra a palabras
                palabra ← cadena vacía
        Sino
            Agregar carácter a palabra

    Si palabras está vacía
        Retornar lista vacía

    // Obtener líneas candidatas usando el índice
    Si solo hay una palabra
        lineas_candidatas ← indexador.obtener_lineas(primera palabra)
    Sino
        lineas_candidatas ← indexador.obtener_lineas(primera palabra)
        Para cada palabra adicional en palabras
            siguientes ← indexador.obtener_lineas(palabra actual)
            lineas_candidatas ← intersección(lineas_candidatas, siguientes)

    resultado ← lista vacía

    // Buscar coincidencias exactas del patrón en las líneas candidatas
    Para cada línea_id en lineas_candidatas
        linea ← texto[línea_id]
        Para cada posición posible en la línea
            coincide ← Verdadero
            Para cada carácter en patrón
                Si el carácter en la línea no coincide con el del patrón
                    coincide ← Falso
                    Salir del ciclo interno
            Si coincide
                antes ← (posición es inicio) O es_separador(caracter anterior)
                después ← (posición es final) O es_separador(caracter siguiente)
                Si antes Y después
                    Agregar (línea_id + 1, posición) a resultado

    Ordenar resultado
    Retornar resultado
FinFunción
</pre>

<h3 style="text-align: center;">Tabla Hash</h3>


La **Clase TablaHash** implementa una estructura de datos de tabla hash con direccionamiento abierto usando listas para manejar colisiones. Al inicializar, crea un arreglo de cubetas vacías de tamaño fijo (por defecto 10007, un número primo para mejorar la dispersión de claves). Su función de hash convierte una cadena clave en un índice usando una combinación lineal y módulo con la capacidad, asegurando un reparto uniforme de las claves.

Para almacenar datos, el método `insertar` busca la cubeta correspondiente y añade el valor a la lista asociada a la clave, evitando duplicados. El método `obtener` permite recuperar rápidamente todos los valores asociados a una clave, mientras que el método `claves` devuelve una lista de todas las claves almacenadas en la tabla. Este diseño permite realizar búsquedas e inserciones eficientes, incluso en presencia de colisiones.
<pre>
Clase TablaHash:
    Inicializar(capacidad = 10007):
        capacidad ← valor dado (por defecto 10007)
        tabla ← lista de listas vacías de tamaño capacidad

    Función _hash(clave):
        h ← 0
        Para cada carácter c en clave:
            h ← (h * 31 + valor_ASCII(c)) módulo capacidad
        Retornar h

    Función insertar(clave, valor):
        pos ← _hash(clave)
        Para cada par en tabla[pos]:
            Si par[0] = clave:
                Si valor no está en par[1]:
                    agregar valor a par[1]
                Retornar
        agregar [clave, [valor]] a tabla[pos]

    Función obtener(clave):
        pos ← _hash(clave)
        Para cada par en tabla[pos]:
            Si par[0] = clave:
                Retornar par[1]
        Retornar lista vacía

    Función claves():
        todas ← lista vacía
        Para cada cubeta en tabla:
            Para cada par en cubeta:
                agregar par[0] a todas
        Retornar todas
</pre>

<h3 style="text-align: center;">Indexador</h3>
La clase Indexador permite construir un índice invertido de palabras para un texto, usando una tabla hash. Al indexar, recorre cada línea, separa y limpia las palabras, y registra en qué líneas aparece cada una. El método limpiar_palabra elimina signos al inicio y final de cada palabra, asegurando que solo se almacenen términos relevantes. Finalmente, el método obtener_lineas permite recuperar rápidamente todas las líneas donde aparece una palabra dada, optimizando la búsqueda de palabras exactas en grandes volúmenes de texto.

<pre>
Clase Indexador:
    Inicializar():
        Crear una tabla hash vacía para el índice de palabras

    Función es_alfanumerico(caracter):
        Retornar True si el carácter es una letra (mayúscula/minúscula, acentuada o ñ) o un número

    Función limpiar_palabra(palabra):
        inicio ← 0
        fin ← longitud de palabra - 1
        Mientras inicio <= fin y palabra[inicio] no es alfanumérico:
            inicio ← inicio + 1
        Mientras fin >= inicio y palabra[fin] no es alfanumérico:
            fin ← fin - 1
        Retornar subcadena de palabra desde inicio hasta fin (inclusive)

    Función indexar(texto):
        Para cada línea con índice i en texto:
            palabra ← cadena vacía
            Para cada carácter j en la línea (incluyendo un carácter final extra):
                Si es separador (espacio, salto de línea, tabulador) o final de línea:
                    Si palabra no vacía:
                        palabra_limpia ← limpiar_palabra(palabra)
                        Si palabra_limpia no vacía:
                            Insertar (palabra_limpia, i) en el índice
                        palabra ← cadena vacía
                Si no:
                    Agregar carácter a palabra

    Función obtener_lineas(palabra):
        palabra_limpia ← limpiar_palabra(palabra)
        Retornar lista de líneas asociadas a palabra_limpia en el índice
</pre>

## 3.  Pruebas y resultados
Se realizaron pruebas de rendimiento para comparar la eficiencia de la búsqueda indexada frente a la búsqueda por fuerza bruta. En cada caso se recopilaron métricas como el tiempo de respuesta y el comportamiento del algoritmo, estableciendo una base sólida para el análisis comparativo de los resultados.
### 3.1  Comparación de Rendimiento por Frecuencia de Palabras

El gráfico presentado compara el rendimiento de la búsqueda por fuerza bruta frente a la búsqueda indexada, en función de la frecuencia de aparición de las palabras en el corpus analizado.

En la parte superior, la gráfica de barras muestra los tiempos promedio de búsqueda (en milisegundos, con escala logarítmica) para ambas técnicas, clasificados en cuatro categorías: palabras raras, poco comunes, comunes y muy comunes. Se observa que la fuerza bruta mantiene tiempos elevados y prácticamente constantes (entre 161 ms y 165 ms), sin importar la frecuencia del término. En contraste, la búsqueda indexada reduce drásticamente los tiempos, manteniéndose siempre por debajo de los 0.05 ms. Esto evidencia que la eficiencia del método indexado es independiente de la frecuencia, a diferencia de la fuerza bruta, cuyo rendimiento no varía significativamente.

En la parte inferior se presenta el factor de aceleración (speedup), obtenido al comparar ambas técnicas. En todos los casos, la búsqueda indexada resulta miles de veces más rápida, alcanzando hasta 8755× para palabras raras. Incluso en el peor caso (palabras muy comunes), se logra una aceleración superior a 4000×. La línea roja horizontal indica el umbral sin mejora (1×), claramente superado en todos los escenarios.

En conjunto, ambos gráficos demuestran que la búsqueda indexada proporciona mejoras de rendimiento masivas, consistentes y sostenidas, lo que la convierte en una solución eficiente, robusta y escalable frente al enfoque tradicional por fuerza bruta.
